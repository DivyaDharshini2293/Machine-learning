```python
import pandas as pd
import numpy as np
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
train=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
train.set_index("PassengerId",inplace=True)
test.set_index("PassengerId",inplace=True)
```

```python
test.head()
```

```python
train.head()
```

```python
train.info()
```

```python
test.info()
```

```python
y_train=train["Survived"]
train.drop("Survived",axis=1,inplace=True)
train.shape
```

```python
train.columns
```

```python
train_test=train.append(test)
```

```python
data=train_test.drop(['Name','Age', 'SibSp', 'Parch', 'Ticket','Cabin', 'Embarked'],axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.fillna(method='ffill',inplace=True)
```

```python
data.isnull().sum()
```

```python
data=pd.get_dummies(data,columns=["Sex"])
data
```

```python
x_train=data.values[:891]
x_test=data.values[891:]
```

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x_train_scale=sc.fit_transform(x_train)
```

```python
x_test_scale=sc.fit_transform(x_test)
```

```python
x_test.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x_train_scale,y_train,test_size=0.2,random_state=42)
X_train.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
learning_rate=[0.01,0.05,0.1,0.25,0.5,0.75,1]
for lr in learning_rate:
  gb=GradientBoostingClassifier(n_estimators=150,learning_rate=lr,max_features=3,max_depth=20)
  gb.fit(X_train,Y_train)
  print("\n")
  print("Learning Rate= ",lr)
  print("training Score ",gb.score(X_train,Y_train))
  print("Test score ",gb.score(X_test,Y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(n_estimators=100,random_state=42)
ad.fit(X_train,Y_train)
print("training Score ",ad.score(X_train,Y_train))
print("Test score ",ad.score(X_test,Y_test))
```

```python
y_pred=ad.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,Y_test)
```
