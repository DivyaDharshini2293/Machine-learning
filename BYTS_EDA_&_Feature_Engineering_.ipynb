#Identifying Outliers


*   Z-Score[z-score= (x-mean)/std deviation]
*   Inter Quartile Range



```python
#z-score
from scipy import stats
z_score=np.abs(stats.zscore(data['normalized-losses']))
z_score
```

```python
data['normalized-losses']
```

```python
import seaborn as sns
sns.boxplot(z_score)
```

```python
threshold=z_score.max()
```

```python
index_val=np.where(z_score >=threshold)
data.iloc[index_val]['normalized-losses']
```

```python
new_data=data.drop(index=data.iloc[index_val]['normalized-losses'].index)
new_data
```
